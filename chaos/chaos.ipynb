```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

s, b, r = 15, 3, 30
vector =  (1, 1, 1)

def lorenz(vector, t, s, r, b):
    x, y, z = vector
    dx = s * (y -x)
    dy = x * (r-z) - y
    dz = x*y - b*z
    return dx, dy, dz

m, n = 150, 100000

t = np.linspace(0, m, n)
f = odeint(lorenz, vector, t, args=(s, r, b))
x, y, z = f.T

figure = plt.figure()
ax = figure.gca(projection='3d')

o = 100
space = np.linspace(0,1,n)
for i in range(0,n-o,o):
    ax.plot(x[i:i+o+1], y[i:i+o+1], z[i:i+o+1], color=(0.8, space[i], 0), alpha=0.8)

plt.show()
```
